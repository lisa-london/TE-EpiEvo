```r
options(repr.plot.width = 5, repr.plot.height = 10)
myplot <- ggplot(tcmar_nodes,aes(DDE3_motif,Degree))+
            geom_boxplot(aes(fill=DDE3_motif,color=DDE3_motif),alpha=0.3)+
           stat_compare_means()+
            theme_light(base_size=25)+
            theme(legend.position='none',text=element_text(size=15,family="Times New Roman"))+
            scale_fill_manual(values=c('#ff7f00','#377eb8'))+
            scale_color_manual(values=c('#ff7f00','#377eb8'))+xlab("")+
            theme(panel.grid.major = element_blank(), panel.grid.minor = element_blank())
            

myplot
```
