```r
library(tidyverse)
library(ggsignif)
library(ggpubr)
require(extrafont)
```

```r
##### loading the node data frame from the insect graph
nodes <- read.csv("EV2.csv")
head(nodes,3)
```

```r
options(repr.plot.width = 5, repr.plot.height = 10)


dde3insect <- ggplot(nodes,aes(DDE3,Degree))+
    geom_boxplot(aes(fill=DDE3,color=DDE3),alpha=0.5)+ 
    stat_compare_means()+
    theme_light(base_size=25)+
    theme(legend.position='none',text=element_text(size=15,family="Times New Roman"))+
    scale_fill_manual(values=c('#ff7f00','#377eb8'))+
    scale_color_manual(values=c('#ff7f00','#377eb8'))

dde3insect
```
